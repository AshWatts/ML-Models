```python
X = data.drop(['species'], axis=1)
Y = data['species']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
print("X Train Shape: ", X_train.shape)
print("X Test Shape: ", X_test.shape)
```

```python
print("Y Train Shape: ", Y_train.shape)
print("Y Test Shape: ", Y_test.shape)
```

```python
X_train.isnull().sum(), X_test.isnull().sum()
```

```python
for col in X_train.columns:
    col_median = X_train[col].median()
    X_train[col] = X_train[col].fillna(col_median)
    X_test[col] = X_test[col].fillna(col_median)
```

```python
X_train.isnull().sum(), X_test.isnull().sum()
```

```python
cols = X_train.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns=[cols])
```

```python
X_test = pd.DataFrame(X_test, columns=[cols])
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7)
# fit the model to the training set
knn.fit(X_train, Y_train)
```

```python
Y_pred=knn.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import accuracy_score
print('Model accuracy score for Test Data Set: {0:0.4f}'. format(accuracy_score(Y_test, Y_pred)))
```

```python
Y_pred_train=knn.predict(X_train)
print('Model accuracy score for Training Data Set: {0:0.4f}'. format(accuracy_score(Y_train, Y_pred_train)))
```

```python
accuracies = []

# Range of k values to try
k_values = range(1, 21)

# Loop over different k values
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, Y_train)
    Y_pred = knn.predict(X_test)
    accuracy = accuracy_score(Y_test, Y_pred)
    accuracies.append(accuracy)

# Find the value of k with the maximum accuracy
max_accuracy = max(accuracies)
best_k = k_values[accuracies.index(max_accuracy)]

# Plot the accuracy for different k values
plt.figure(figsize=(12, 6))
plt.plot(k_values, accuracies, marker='o', color='b', linestyle='-', linewidth=2, markersize=8, label='Accuracy')

# Highlight the best k value
plt.axvline(x=best_k, color='r', linestyle='--', label=f'Best k = {best_k}')
plt.scatter(best_k, max_accuracy, color='red', s=100, zorder=5)
plt.text(best_k, max_accuracy, f'{best_k}: {max_accuracy:.2f}', horizontalalignment='right')

# Add titles and labels
plt.title('Accuracy of k-NN for Different Values of k', fontsize=16)
plt.xlabel('Number of Neighbors (k)', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)

# Customize ticks
plt.xticks(k_values, fontsize=12)
plt.yticks(np.arange(0.8, 1.05, 0.02), fontsize=12)

# Add grid
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Add legend
plt.legend(fontsize=12)

# Show the plot
plt.show()
```

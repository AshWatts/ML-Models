```python
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings = pd.read_csv("toy_dataset.csv", index_col=0)
ratings = ratings.fillna(0)
ratings
```

```python
def standardize(row):
    new_row = (row - row.mean()) / (row.max() - row.min())
    return new_row
```

```python
ratings_std = ratings.apply(standardize)
ratings_std
```

```python
item_similarity = cosine_similarity(ratings_std.T)
print(item_similarity)
```

```python
item_similarity_df = pd.DataFrame(item_similarity, index=ratings.columns, columns=ratings.columns)
item_similarity_df
```

```python
def get_similar_movies(movie, rating):
    # Example DataFrame for demonstration purposes
    data = {
        'Movie': [f'{movie}_1', f'{movie}_2', f'{movie}_3'],
        'Rating': [rating + i for i in range(3)]
    }
    return pd.DataFrame(data)

# List of tuples with movie and rating
action_lover = [("action1", 5), ("romantic2", 1), ("romantic3", 1)]

# Initialize a list to collect DataFrames
similar_movies_list = []

# Iterate over the action_lover list and get DataFrames
for movie, rating in action_lover:
    df = get_similar_movies(movie, rating)
    similar_movies_list.append(df)

# Concatenate DataFrames vertically
similar_movies = pd.concat(similar_movies_list, ignore_index=True)

# Display the resulting DataFrame
print("DataFrame:\n", similar_movies)

# Calculate and display the sum
print("\nSum:\n", similar_movies.sum(numeric_only=True))
```
